```python
def read_position(df, row, col):
    return df.iat[row, col]
```

```python
def summarize_attribute_distribution(df):
    """
    Prints the distribution of each attribute in the DataFrame.
    For numerical attributes: prints basic statistics.
    For categorical attributes: prints value counts.
    """
    for column in df.columns:
        print(f"\n=== Attribute: {column} ===")

        # Numeric distribution
        if df[column].dtype in ['int64', 'float64']:
            print("Numeric distribution:")
            print(df[column].describe())

        # Categorical / text distribution
        else:
            print("Value frequencies:")
            print(df[column].value_counts())
```

```python
def summarize_attribute_distribution(df):
    """
    Prints the describe() table for the whole DataFrame.
    Then prints value counts ONLY for columns where count != unique.
    """
    # Always show the describe() table first
    print(" Describe() summary for all attributes ")
    display(df.describe(include='all'))
    
    # Then additionally show value counts for relevant columns
    for column in df.columns:
        desc = df[column].describe()

        # Only print value counts when values repeat
        if 'unique' in desc and desc['count'] != desc['unique']:
            print(f"\n Value counts for: {column}")
            print(df[column].value_counts())
```

```python
def summarize_missing_data(df):
    """
   
    - Attributes (columns) that contain missing values
    - Entries (rows) that contain missing values
    """

    # Missing values per attribute
    missing_per_attribute = df.isnull().sum()
    missing_per_attribute = missing_per_attribute[missing_per_attribute > 0]
    missing_per_attribute = missing_per_attribute.to_frame(name="missing_count")
    missing_per_attribute.index.name = "attribute"

    # Missing values per entry (row)
    missing_per_entry = df.isnull().sum(axis=1)
    missing_per_entry = missing_per_entry[missing_per_entry > 0]
    missing_per_entry = missing_per_entry.to_frame(name="missing_count")
    missing_per_entry.index.name = "row"

    return missing_per_attribute, missing_per_entry

     
```

```python
attr_missing, row_missing = summarize_missing_data(df_HBM_services_weekly)

display(attr_missing)
display(row_missing)
```
